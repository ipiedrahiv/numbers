# Taller Práctico Funciones

## Punto 1

Considere la siguiente función

$ f(x) = e^{x+3} + \frac{x^3}{3} - 10 $

1. Demuestre que la función tiene una raíz única en el intervalo $ [-1, 0] $.
2. Justifique que se puede aplicar el método de la bisección en $ [-1, 0] $ para aproximar la raíz de la ecuación $ f(x) = 0 $ y luego úselo para aproximar la raíz con una tolerancia de $ 10^{-6} $.
3. ¿Cuantas iteraciones serían necesarias para obtener un eeror absoluto menos a $ 10^{-8} $?

```python
# Hacemos la función

def f1(x):
    return math.exp(x + 3) + (math.pow(x, 3) / 3) - 10
```

```python
a = -1
b = 0

# Verificamos que la raíz equite y es única

# Verificar si f es continua en [a, b]

"""No hay logaritmos, raices cuadradas o fracciones que puedan tener divisores iguales a 0, por lo que podemos asumir que la función es 
continua en todo su dominio, pero podemos graficar para estar más seguros de esta intuición."""

# Verificar que f(a) y f(b) tengan signos opestos

print(f1(a)*f1(b))

# Verificar que f(x) es diferenciable en (a,b)y que f'(x) no cambia de signo en [a,b]

"""Notamos que en el intervalo (a, b) no hay picos ni discontinuidades, por lo que x es 
diferenciable en todo el intervalo. Adicionalmente no hay ningún máximo o mínimo local dentro 
del intervalo, por lo que podemos concluir que la derivada no cambia de signo."""

# Concluiomos que existe una raíz única ya que se cumplen los tres requisitos.
```

```python
"""Se puede aplicar el m;etodo de la bisección al intervalo [a, b] ya que se cumple el teorema
del valor medio y sabemos a ciencia cierta que ahí hay una raiz."""

def bisection_method(f, a, b, tolerance):
    # We calculate the max number of iterations based on the tolerance
    max_iterations = math.ceil(math.log((b-a) * (1/tolerance)) / math.log(2))
    
    # We calculate the midpoint of the given interval
    c = (a+b)/2
    previous_c = c # This is an auxiliar variable that will allow us to save the c of the last iteration, it will be used to calculat the error
    
    # We evaluate all the relevant points of the interval on the given function
    f_a = f(a)
    f_b = f(b)
    f_c = f(c)
    
    # We initializa a counter
    count = 0
    
    # We initialize an empty variable in which we will store the absolute error
    err = None
    
    # We create a data frame in which we will sotre the obtained values, to create the method chart
    pd.set_option("display.precision", 8)
    data = {'n': [count],
            'a': [a],
            'c': [c],
            'b': [b],
            'f(c)': [f_c],
            'E': [err]}
    output = pd.DataFrame(data)

    # We start iterating, the tolerance will not be considered as a condition because we already calculated the amount of iterations necessary to achieve the decired precition
    while (count < max_iterations):
        # Is the root in the left interval?
        if f_a * f_c < 0:
            b = c
            f_b = f(b)
        # Is it in the right interval?    
        elif f_b * f_c < 0:
            a = c
            f_a = f(a)
        
        # Calculate new c
        c = (a+b)/2
        
        # Calculate absolute error
        err = abs(previous_c - c)
        
        # Evaluate the function in the new c value
        f_c = f(c)
        
        # Add one to the counter
        count += 1
        
        # Append the new row to the method chart
        new_line = {'n':count,
                    'a': a,
                    'c':c,
                    'b': b,
                    'f(c)':f_c,
                    'E':err}
        output = output.append(new_line, ignore_index=True)
        
        # Sace the current c as previous c, in order to calculate E later on
        previous_c = c
        
        
    print(output)
    print (f'There is a root in the interval [{a},{b}]')
    return [a, b]

bisection_method(f1, -1, 0, 0.000001)
```

```python
# Usamos el teorema de error del método de bisección y despejamos para hallar la siguiente función

print(math.log(1, 2) - math.log(math.pow(10, -8), 2)/math.log(2, 2))

"""Se necesitarán 27 iteraciones para alcanzar un error menos a 0.00000001"""
```

## Punto 2

Considere la siguiente función:

$ f(x) = x^2 - 2x + 2 - e^x $

1. Compruebe que el intervalo [0.1, 0.7] solo teiene una raíz real.
2. Para cada una de las siguientes funciones dadas demuestre que la ecuación $f(x) = 0$ esequivalente algebraicamente (en cierto dominio) a la ecuación $x=g(x)$ 
    
    $g_1(x) = ln(x^2 - 2x + 2)$
    
    $g_2(x) = \sqrt{e^x + 2x - 2}$
    
    $g_3(x) = \frac{x^2+ 2 - e^x}{2}$
    
3. Pruebe que $g_1(x)$ satisface las hipótesis del teorema de punto fijo en [0:1;0:7] y aplique el método de punto fijo con la función de iteración $g_1(x)$ para aproximar la única raíz real de f(x) = 0; use p0 = 0.5 y una tolerancia para el error relativo de $10^{-8}$.


1. Sabemos que la función es continua ya que no tiene posibles asíntotas, no hay logaritmos, ni gracciones ni raices cuadradas. Por lo que podemos comprobar que es muy poco probable que exista una discontinuidad. Esto se evidencia en la gráfica.
2. Podemos ver que $g_1(x)$, $g_2(x)$ y $g_3(x)$ son algebraicamente equivalentes a $f(x)$ ya que todas estas se pueden conseguir al despejar alguna de las $x$ de $f(x)$. La primera se consigue despejando la $x$ en el término del exponenete. La segunda viene de despejar la $x$ que está elevada al cuadrado. La tercera viene de despejar la que está en el factor $2x$. Por lo que todas son algebraicamente equivalentes a la original.



```python
# General chart structure: n, xn, checking condition, error 

def fixed_point(f, g, x0, tolerance, max_iterations):
    current_aproximation = x0
    current_f = f(current_aproximation) 
    n = 0
    
    # We create a data frame in which we will sotre the obtained values, to create the method chart
    pd.set_option("display.precision", 10)
    data = {'n': [n],
            'Xn': [current_aproximation],
            'f(x)': [current_f],
            'E': [None]}
    output = pd.DataFrame(data)
    
    while (n < max_iterations):
        n += 1
        
        current_f = f(current_aproximation)
        previous_aproximation = current_aproximation
        current_aproximation = g(previous_aproximation)
        err = abs(current_aproximation - previous_aproximation)
        
        new_line = {'n':n,
                    'Xn': current_aproximation,
                    'f(x)':current_f,
                    'E':err}
        output = output.append(new_line, ignore_index=True)
        
        if err <= tolerance:
            break
            
    print(output)
```

```python
def g2(x):
    return math.log(math.pow(x, 2)-(2*x)+(2))

def f2(x):
    return math.pow(x, 2) - (2*x) + (2) - math.exp(x)
```

```python
fixed_point(f2, g2, 0.5, 0.00000001, 200)
```

## Punto 6

Considere el polinomio $p(x) = 7.14292992 - 20.0471x + 19.0956x^2 - 7.45x^3 + x^4 $ del cual se sabe que contiene una raíz múltiple.

1. Utilize el método de las raices múltiples.
2. Use el método Newton-Raphson acelerado

```python
def multiple(f, f_derivate, f_second_derivate, x0, tolerance, max_iterations):
    n = 0
    
    # NOOOO
    alpha = 2
    # NOOOO
    
    pd.set_option("display.precision", 6)
    data = {'n': [n],
            'Xn': [x0],
            'f(x)': [f(x0)],
            'f\'(x)': [f_derivate(x0)],
            'f\'\'(x)':[f_second_derivate(x0)],
            'E': [None],}
    output = pd.DataFrame(data)
    
    while n < max_iterations:
        n += 1
        
        x1 = x0 - ((f(x0)*f_derivate(x0))/(math.pow(f_derivate(x0),2)-(f(x0)*f_second_derivate(x0))))
        err = abs(x1 - x0)
        x0 = x1
        
        new_line = {'n':n,
                    'Xn': x1,
                    'f(x)':f(x1),
                    'f\'(x)': f_derivate(x1),
                    'f\'\'(x)':f_second_derivate(x1),
                    'E':err,}
        output = output.append(new_line, ignore_index=True)
        
        print(output.loc[n,"E"]/math.pow(output.loc[n-1,"E"],alpha))
        
        if err <= tolerance:
            break
            
        
        
    print (output)
```

```python
def f3(x):
    return 7.14292992 - 20.0471*x + 19.0956*math.pow(x,2) - 7.45*math.pow(x, 3) + math.pow(x, 3)

def f3_first(x):
    return 4*math.pow(x, 3) - 22.35*math.pow(x, 2) + 38.1912*x - 20.0471

def f3_second(x):
    return 12*math.pow(x, 2) - 44.7*x + 38.1912
```

```python
multiple(f3, f3_first, f3_second, 1.4,  0.00001, 10000)
```

```python
def newton_acelerado(f, f_derivate, M, x0, tolerance, max_iterations):
    n = 0
    
    pd.set_option("display.precision", 8)
    data = {'n': [n],
            'Xn': [x0],
            'f(x)': [f(x0)],
            'f\'(x)': [f_derivate(x0)],
            'E': [None]}
    output = pd.DataFrame(data)
    
    while n < max_iterations:
        if f_derivate(x0) == 0:
            return None
        
        n += 1
        
        x1 = x0 - M*(f(x0) / f_derivate(x0))
        err = abs(x1 - x0)
        x0 = x1
        
        new_line = {'n':n,
                    'Xn': x1,
                    'f(x)':f(x1),
                    'f\'(x)': f_derivate(x1),
                    'E':err}
        output = output.append(new_line, ignore_index=True)
        
        if err <= tolerance:
            break
            
        
        
    print (output)
```

```python
newton_acelerado(f3, f3_first, 2, 1.4, 0.00001, 100000)
```
